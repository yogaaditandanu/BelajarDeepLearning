```python
import pandas as pd
import tensorflow as tf
df = pd.read_csv('https://drive.google.com/uc?id=15RfMD9lNkpS3cVN7j3_dsJKZ8_5RJG5z')
df.head()
```

```python
def windowed_dataset(series, window_size, batch_size, shuffle_buffer): #definisi fungsi yg menerima 4 parameter,
#series : data deret waktu, window_size = ukuran jendela yg akan digunakan, batch size = ukuran batch, shuffle buffer = ukuran utk pengacakan
    series = tf.expand_dims(series, axis=-1) #mengubah dimensi series dgn menambahkan 1 dimensi baru di posisi terakhir
    ds = tf.data.Dataset.from_tensor_slices(series) #membuat dataset TF menjadi sebuah series
    ds = ds.window(window_size + 1, shift=1, drop_remainder=True) #membagi dataset menjadi jendela berukuran window_size+1 dgn jarak 1 antara jendela tsb
    ds = ds.flat_map(lambda w: w.batch(window_size + 1)) #mengubah setiap window menjadi sebuah batch dgn ukuran window_size+1
    ds = ds.shuffle(shuffle_buffer) #mengacak dataset dgn menggunakan buffer ukuran shuffle_buffer
    ds = ds.map(lambda w: (w[:-1], w[-1:])) #memetakan setiap batch menjadi sepasang data input & target
    return ds.batch(batch_size).prefetch(1) #mengelompokkan batch dataset menjadi batch dgn ukuran batch size
```

```python
dates = df['date'].values
temp = df['meantemp'].values
```

Kode di atas akan mengambil seluruh data dari dataset yang sebelumnya sudah kita siapkan.

```python
train_set = windowed_dataset(temp, window_size=2, batch_size=3, shuffle_buffer=1000)
```
